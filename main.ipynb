```python
# for each row in th, if value is 1, then mark the corresponding row in matrix as 1
def mark_theory_slots_in_matrix(th, matrix):
    for i in range(len(th)):
        for j in range(len(th.columns)):
            if th.iloc[i, j] == 1:
                matrix[i, j, :] = 1
    return matrix

matrix = mark_theory_slots_in_matrix(th, matrix)
matrix

# If the row is 1 in matrix, then mark the last cell of the row as 0
def mark_0_in_last_cell(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if matrix[i, j, 0] == 1:
                matrix[i, j, 6] = 0
    return matrix

matrix = mark_0_in_last_cell(matrix)
matrix
```
